```python
!pip install ultralytics

from ultralytics import YOLO

# Load the model (same as before)
# model = YOLO("yolo11l.pt")
model = YOLO("/kaggle/input/vehicle-dataset/best.pt")  # using a custom model

# Validate the model on GPU and with test split
metrics = model.val(device=0, data='/kaggle/input/vehicle-dataset/config3.yaml', plots=True, save_json=True)  

# Print performance metrics
print("mAP50-95:", metrics.box.map)  # map50-95
print("mAP50:", metrics.box.map50)  # map50
print("mAP75:", metrics.box.map75)  # map75
print("mAPs:", metrics.box.maps)  # a list contains map50-95 of each category
```
